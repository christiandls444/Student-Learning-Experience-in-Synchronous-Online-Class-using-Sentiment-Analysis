```python
from PIL import Image, ImageTk
from tkinter import ttk
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from tkinter import *
from openpyxl import *


wb = load_workbook('dataset.xlsx')
sheet = wb.active


def excel():
    sheet.column_dimensions['A'].width = 30
    sheet.column_dimensions['B'].width = 15
    sheet.column_dimensions['C'].width = 15
    sheet.column_dimensions['D'].width = 15
    sheet.column_dimensions['E'].width = 15
    sheet.column_dimensions['F'].width = 30
    sheet.column_dimensions['G'].width = 50
    sheet.column_dimensions['H'].width = 20
    
    sheet.cell(row=1, column=1).value = "FULLNAME"
    sheet.cell(row=1, column=2).value = "STUDENT NO"
    sheet.cell(row=1, column=3).value = "COURSE"
    sheet.cell(row=1, column=4).value = "GENDER"
    sheet.cell(row=1, column=5).value = "SECTION"
    sheet.cell(row=1, column=6).value = "EMAIL"
    sheet.cell(row=1, column=7).value = "COMMENT"
    sheet.cell(row=1, column=8).value = "SENTIMENT"


def focus1(event):
    course_field.focus_set()

    
def focus2(event):
    sem_field.focus_set()


def focus3(event):
    form_no_field.focus_set()
    
    
def focus4(event):
    contact_no_field.focus_set()


def focus5(event):
    email_id_field.focus_set()


def focus6(event):
    textArea.focus_set()

    
def focus7(event):
    overallField.focus_set()


def clearAll():
    negativeField.delete(0, END)
    neutralField.delete(0, END)
    positiveField.delete(0, END)
    overallField.delete(0, END)
    textArea.delete(0, END)
    name_field.delete(0, END)
    course_field.delete(0, END)
    sem_field.delete(0, END)
    form_no_field.delete(0, END)
    contact_no_field.delete(0, END)
    email_id_field.delete(0, END)


def insert():
    if (name_field.get() == "" and
        course_field.get() == "" and
        sem_field.get() == "" and
        form_no_field.get() == "" and
        contact_no_field.get() == "" and
        email_id_field.get() == "" and
        textArea.get() == ""):       
        print("empty input")
    else:
        current_row = sheet.max_row
        current_column = sheet.max_column
        
        sheet.cell(row=current_row + 1, column=1).value = name_field.get()
        sheet.cell(row=current_row + 1, column=2).value = course_field.get()
        sheet.cell(row=current_row + 1, column=3).value = sem_field.get()
        sheet.cell(row=current_row + 1, column=4).value = form_no_field.get()
        sheet.cell(row=current_row + 1, column=5).value = contact_no_field.get()
        sheet.cell(row=current_row + 1, column=6).value = email_id_field.get()
        sheet.cell(row=current_row + 1, column=7).value = textArea.get()
        sheet.cell(row=current_row + 1, column=8).value = overallField.get()
  
        wb.save('dataset.xlsx')
        name_field.focus_set()
        clearAll()


def detect_sentiment():
    sentence = textArea.get()
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)

    string = str(sentiment_dict['neg']*100) 
    negativeField.insert(10, string)

    string = str(sentiment_dict['neu']*100)
    neutralField.insert(10, string)

    string = str(sentiment_dict['pos']*100) 
    positiveField.insert(10, string)

    if sentiment_dict['compound'] >= 0.05:
        string = "1"
    elif sentiment_dict['compound'] <= - 0.05:
        string = "0"
    else:
        string = "2"

    overallField.insert(10, string)


if __name__ == "__main__":
    gui = Tk()
    img= PhotoImage(file='star.png', master= gui)
    img_label= Label(gui,image=img)
    img_label.place(x=-10, y=-350)
    gui.title("SLESOC using Sentiment Analysis")
    
    w = 485
    h = 820
    ws = gui.winfo_screenwidth()
    hs = gui.winfo_screenheight()
    x = (ws/2) - (w/2)
    y = (hs/2) - (h/2)
    gui.geometry('%dx%d+%d+%d' % (w, h, x, y))
    
    excel()
  
    name = Label(gui, text="FULLNAME",bg="#000002" , fg="white",font=('Century', 12, 'bold'))  
    course = Label(gui, text="STUDENT NO",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    sem = Label(gui, text="COURSE",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    form_no = Label(gui, text="GENDER",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    contact_no = Label(gui, text="SECTION",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    email_id = Label(gui, text="EMAIL",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    
    enterText = Label(gui, text="COMMENT",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    textArea = Text(gui, height=5, width=25, font="lucida 13")
    check = Button(gui, text="CHECK SENTIMENT", fg="white",bg="#ff1414", command=detect_sentiment,font=('Century', 12, 'bold'))
    negative = Label(gui, text="NEGATIVE RATED AS",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    neutral = Label(gui, text="NEUTRAL RATED AS",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    positive = Label(gui, text="POSITIVE RATED AS",bg="#000002" , fg="white",font=('Century', 12, 'bold'))
    overall = Label(gui, text="OVERALL RATED AS",bg="#000002", fg="white",font=('Century', 12, 'bold'))
    
    name_field = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    course_field = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    sem_field = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    form_no_field = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    contact_no_field = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    email_id_field = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    textArea = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    
    negativeField = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    neutralField = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    positiveField = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12))
    overallField = Entry(gui, justify=CENTER, fg="#262D35",font=("Century", 12,'bold'))
    clear = Button(gui, text="CLEAR ALL", fg="white",bg="#ff1414", command=clearAll,font=('Century', 12, 'bold'))
    Exit = Button(gui, text="SUBMIT", fg="white",bg="#ff1414", command=insert,font=('Century', 12, 'bold'))

    
    name_field.bind("<Return>", focus1)
    course_field.bind("<Return>", focus2)
    sem_field.bind("<Return>", focus3)
    form_no_field.bind("<Return>", focus4)
    contact_no_field.bind("<Return>", focus5)
    email_id_field.bind("<Return>", focus6)
    
    
    name.grid(row=0, column=2, ipadx="78", ipady="2")
    name_field.grid(row=1, column=2,ipadx="150", ipady="2")
    
    course.grid(row=2, column=2,ipadx="70", ipady="2")
    course_field.grid(row=3, column=2,ipadx="150", ipady="2")
    
    sem.grid(row=4, column=2,ipadx="91", ipady="2")
    sem_field.grid(row=5, column=2,ipadx="150", ipady="2")
    
    form_no.grid(row=6, column=2,ipadx="103", ipady="2")
    form_no_field.grid(row=7, column=2,ipadx="150", ipady="2")
    
    contact_no.grid(row=8, column=2,ipadx="86", ipady="2")
    contact_no_field.grid(row=9, column=2,ipadx="150", ipady="2")
    
    email_id.grid(row=10, column=2,ipadx="97", ipady="2")
    email_id_field.grid(row=11, column=2,ipadx="150", ipady="2")

    
    enterText.grid(row=12, column=2,ipadx="79", ipady="2")
    textArea.grid(row=13, column=2,ipadx="150", ipady="2")
    check.grid(row=14, column=2,ipadx="31", ipady="2")
    negative.grid(row=15, column=2,ipadx="29", ipady="2")
    negativeField.grid(row=16, column=2,ipadx="150", ipady="2")
    neutral.grid(row=17, column=2,ipadx="29", ipady="2")
    neutralField.grid(row=18, column=2,ipadx="150", ipady="2")
    positive.grid(row=19, column=2,ipadx="29", ipady="2")
    positiveField.grid(row=20, column=2,ipadx="150", ipady="2")
    overall.grid(row=21, column=2,ipadx="29", ipady="2")
    overallField.grid(row=22, column=2,ipadx="140", ipady="2")
    
    clear.grid(row=23, column=2,ipadx="64", ipady="2")
    Exit.grid(row=24, column=2,ipadx="80", ipady="2")

    
    gui.mainloop()
```
